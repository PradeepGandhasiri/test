```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('http://www.cricbuzz.com/live-cricket-scorecard/18877/eng-vs-ind-1st-t20i-india-tour-of-england-2018')
```

```python
c=r.content
```

```python
soup=BeautifulSoup(c,'html.parser')
```

```python
table=[]
for m in soup.find_all('div',{'class':'cb-col cb-col-100 cb-ltst-wgt-hdr'}):
    team=''
    for j in m.find_all('div',{'class':'cb-col cb-col-100 cb-scrd-hdr-rw'}):
        try:
            k=j.find('span').text
            n=k.split(' ')
            team=n[0]
        except:
            pass
    for i in m.find_all('div',{'class':'cb-col cb-col-100 cb-scrd-itms'}):
        d={}
        for j in i.find_all('div',{'class':'cb-col cb-col-27 '}):
            d['Players']=j.text.strip()
        for j in i.find_all('div',{'class':'cb-col cb-col-8 text-right text-bold'}):
            d['Runs']=j.text
        count=0
        for j in i.find_all('div',{'class':'cb-col cb-col-8 text-right'}):
            if count==1:
                d['Fours']=j.text
            elif count==2:
                d['Sixes']=j.text
            elif count==3:
                d['Strike Rate']=j.text
            count+=1
        for j in i.find_all('div',{'cb-col-73 cb-col'}):
            temp=j.text
            li=temp.split(',')
            for k in li:
                e={}
                e['Players']=k.strip()
                e['Runs']=0
                e['Fours']=0
                e['Sixes']=0
                e['Strike Rate']=0.0
                e['Team']=team
                table.append(e)
        if d=={}:
            pass
        else:
            d['Team']=team
            table.append(d)
```

```python
table1=[]
team=[]
count1=1
for m in soup.find_all('div',{'class':'cb-col cb-col-100 cb-ltst-wgt-hdr'}):
    for j in m.find_all('div',{'class':'cb-col cb-col-100 cb-scrd-hdr-rw'}):
        try:
            k=j.find('span').text
            n=k.split(' ')
            if n[0] not in team:
                team.append(n[0])
        except:
            pass
    for i in m.find_all('div',{'class':'cb-col cb-col-100 cb-scrd-itms '}):
        d={}
        for j in i.find_all('div',{'class':'cb-col cb-col-40'}):
            d['Players']=j.text.strip()
        for j in i.find_all('div',{'class':'cb-col cb-col-8 text-right text-bold'}):
            d['Wickets']=j.text
        count=0
        for j in i.find_all('div',{'class':'cb-col cb-col-8 text-right'}):
            if count==0:
                balls=j.text.split('.')
                try:
                    d['Balls']=int(balls[0])*6+int(balls[1])
                except:
                    d['Balls']=int(balls[0])*6
            elif count==1:
                d['Maidens']=j.text
            count+=1
        count=0
        for j in i.find_all('div',{'class':'cb-col cb-col-10 text-right'}):
            if count==0:
                d['Runs']=j.text
            elif count==1:
                d['Economy']=j.text
            count+=1
       
        if d=={}:
            pass
        else:
            table1.append(d)
    count1+=1
```
